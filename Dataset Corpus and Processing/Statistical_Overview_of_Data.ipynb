# Drug Distribution in Training Set

```python
drugbank_e1 = train_set_drugbank["e1_text"].unique()
drugbank_e2 = train_set_drugbank["e2_text"].unique()
drugbank_entities = np.concatenate((drugbank_e1, drugbank_e2))
drugbank_entities
```

```python
len(drugbank_entities)
```

```python
entities = pd.concat([
    train_set_drugbank[['e1_text', 'e1_type']].rename(columns={'e1_text': 'entity', 'e1_type': 'label'}),
    train_set_drugbank[['e2_text', 'e2_type']].rename(columns={'e2_text': 'entity', 'e2_type': 'label'})
])
unique_entities = entities.drop_duplicates()

label_distribution = unique_entities['label'].value_counts()
label_distribution
```

```python
entities = pd.concat([
    train_set_medline[['e1_text', 'e1_type']].rename(columns={'e1_text': 'entity', 'e1_type': 'label'}),
    train_set_medline[['e2_text', 'e2_type']].rename(columns={'e2_text': 'entity', 'e2_type': 'label'})
])
unique_entities = entities.drop_duplicates()

label_distribution = unique_entities['label'].value_counts()
label_distribution
```
